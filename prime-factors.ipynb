```python
"""
Task:
Decompose natural numbers from 1 to N to prime factors
"""

def primefacs(n):
    "Decomposes number to prime factors"
    i = 2
    primefac = []
    while i * i <= n:
        while n % i == 0:
            primefac.append(i)
            n = n // i
        i = i + 1
    primefac.append(n)
    return primefac

if __name__ == "__main__":
    n = int(input("Enter the top boundary: "))
    outputs = []
    for i in range(1, n+1, 1):
        tmp = primefacs(i)
        outputs.append(tmp)
    for i in outputs:
        print(i)
```
